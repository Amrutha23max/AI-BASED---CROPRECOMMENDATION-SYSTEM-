```python
import pandas as pd
df = pd.read_csv('Crop_recommendation.csv')
print(df.shape)
print(df.head())
print(df.columns.tolist())
```

```python
print(f"Total crops: {df['label'].nunique()}")
print(f"\nCrop list:\n{df['label'].value_counts()}")
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

features = ['N','P','K','temperature','humidity','ph','rainfall']

plt.figure(figsize=(16, 10))
for i, col in enumerate(features):
    plt.subplot(2, 4, i+1)
    plt.hist(df[col], bins=30, color='steelblue', edgecolor='black')
    plt.title(col)
plt.suptitle('Feature Distributions', fontsize=14)
plt.tight_layout()
plt.savefig('eda_distributions.png')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(df[features].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('eda_correlation.png')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
df.groupby('label')['N'].mean().sort_values().plot(kind='bar', color='green')
plt.title('Average Nitrogen (N) requirement per Crop')
plt.ylabel('N value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('eda_nitrogen.png')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('Crop_recommendation.csv')
print(df[['N', 'P', 'K', 'ph']].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']

for feature in features:
    plt.figure(figsize=(16, 4))
    sns.boxplot(x='label', y=feature, data=df, palette='Set2')
    plt.title(f'{feature} distribution across crops')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f'eda_boxplot_{feature}.png')
    plt.show()
```
