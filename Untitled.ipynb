```python
from icrawler.builtin import BingImageCrawler
import os, time, math

# Exact targets per class
TARGETS = {
    "Black Soil":    121,
    "Laterite Soil": 242,
    "Yellow Soil":   244,
}

QUERIES = {
    "Black Soil": [
        "black cotton soil texture closeup",
        "black clayey soil sample",
        "black regur soil india",
        "dark black agricultural soil",
        "black soil cracked dry",
    ],
    "Laterite Soil": [
        "laterite soil red texture",
        "laterite soil india closeup",
        "red laterite soil sample",
        "laterite soil agriculture india",
        "laterite rock soil formation",
        "red soil laterite closeup",
        "laterite soil Karnataka Kerala",
    ],
    "Yellow Soil": [
        "yellow soil india closeup",
        "yellow loamy soil sample",
        "yellow sandy soil texture",
        "yellow alluvial soil india",
        "light yellow soil agriculture",
        "yellow soil field india",
        "yellow soil sample close up",
    ],
}

RAW_DIR = "raw"

for soil_class, total_needed in TARGETS.items():
    class_dir = os.path.join(RAW_DIR, soil_class)
    os.makedirs(class_dir, exist_ok=True)

    queries    = QUERIES[soil_class]
    per_query  = math.ceil(total_needed / len(queries))  # spread evenly

    print(f"\n📸 {soil_class} — need {total_needed} more ({per_query}/query × {len(queries)} queries)")

    for query in queries:
        print(f"   🔍 {query}")
        try:
            BingImageCrawler(
                storage={"root_dir": class_dir},
                downloader_threads=4,
                parser_threads=2,
            ).crawl(
                keyword=query,
                max_num=per_query,
                min_size=(150, 150),
                filters={"type": "photo"},
            )
        except Exception as e:
            print(f"   ⚠️ Failed: {e}")
        time.sleep(2)

    count = len([f for f in os.listdir(class_dir)
                 if f.lower().endswith((".jpg",".jpeg",".png",".webp"))])
    print(f"   ✅ {soil_class}: {count} images total")

# ── Final check ──
print("\n── Final counts ──")
for cls in TARGETS:
    count = len([f for f in os.listdir(f"raw/{cls}")
                 if f.lower().endswith((".jpg",".jpeg",".png",".webp"))])
    status = "✅" if count >= 300 else f"⚠️  Still need {300 - count} more"
    print(f"{status}  {cls}: {count} images")
```

```python
import os, shutil, random

RAW_DIR   = "raw"
BASE_DIR  = "dataset"
CLASSES   = ["Black Soil", "Laterite Soil", "Yellow Soil"]
SPLITS    = {"train": 0.70, "valid": 0.15, "test": 0.15}
SEED      = 42

random.seed(SEED)

for cls in CLASSES:
    src   = os.path.join(RAW_DIR, cls)
    imgs  = [f for f in os.listdir(src)
             if f.lower().endswith((".jpg", ".jpeg", ".png", ".webp"))]
    random.shuffle(imgs)

    n       = len(imgs)
    t_end   = int(n * SPLITS["train"])
    v_end   = t_end + int(n * SPLITS["valid"])

    splits = {
        "train": imgs[:t_end],
        "valid": imgs[t_end:v_end],
        "test" : imgs[v_end:]
    }

    for split, files in splits.items():
        dest = os.path.join(BASE_DIR, split, cls)
        os.makedirs(dest, exist_ok=True)
        for f in files:
            shutil.copy(os.path.join(src, f), os.path.join(dest, f))

    print(f"✅ {cls}: {len(splits['train'])} train | "
          f"{len(splits['valid'])} valid | "
          f"{len(splits['test'])} test")

print("\n── Final split counts ──")
for split in ["train", "valid", "test"]:
    total = sum(
        len(os.listdir(os.path.join(BASE_DIR, split, cls)))
        for cls in CLASSES
    )
    print(f"  {split}: {total} images")
print("\n✅ Split complete! Raw originals untouched in raw/")
```

```python
import os, numpy as np
from tensorflow.keras.preprocessing.image import (
    ImageDataGenerator, load_img, img_to_array, save_img
)

TRAIN_DIR = "dataset/train"
AUG_DIR   = "dataset/train_augmented"
CLASSES   = ["Black Soil", "Laterite Soil", "Yellow Soil"]
TARGET_PER_CLASS = 1667   # ~5000 / 3

aug = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.15,
    height_shift_range=0.15,
    horizontal_flip=True,
    vertical_flip=False,
    zoom_range=0.2,
    brightness_range=[0.7, 1.3],
    shear_range=0.1,
    fill_mode="nearest"
)

for cls in CLASSES:
    src_dir  = os.path.join(TRAIN_DIR, cls)
    dest_dir = os.path.join(AUG_DIR, cls)
    os.makedirs(dest_dir, exist_ok=True)

    # First copy all originals into augmented folder
    originals = [f for f in os.listdir(src_dir)
                 if f.lower().endswith((".jpg", ".jpeg", ".png", ".webp"))]

    for f in originals:
        src_path  = os.path.join(src_dir, f)
        dest_path = os.path.join(dest_dir, f)
        img = load_img(src_path, target_size=(224, 224))
        save_img(dest_path, img_to_array(img))

    current_count = len(originals)
    needed        = TARGET_PER_CLASS - current_count
    per_image     = int(np.ceil(needed / current_count))

    print(f"\n📸 {cls}: {current_count} originals → generating {needed} augmented copies")

    aug_count = 0
    for f in originals:
        if aug_count >= needed:
            break

        img_path = os.path.join(src_dir, f)
        img      = load_img(img_path, target_size=(224, 224))
        arr      = img_to_array(img)
        arr      = arr.reshape((1,) + arr.shape)

        for i, batch in enumerate(aug.flow(arr, batch_size=1)):
            if aug_count >= needed:
                break
            fname = f"aug_{cls.replace(' ','_')}_{aug_count:05d}.jpg"
            save_img(os.path.join(dest_dir, fname), batch[0])
            aug_count += 1

    total = len(os.listdir(dest_dir))
    print(f"   ✅ {cls}: {total} images in train_augmented/")

# ── Final count check ──
print("\n── Augmented train counts ──")
grand_total = 0
for cls in CLASSES:
    count = len(os.listdir(os.path.join(AUG_DIR, cls)))
    grand_total += count
    print(f"  {cls}: {count} images")
print(f"\n  Total train images: {grand_total}")
print(f"  Valid images: 63 (unchanged)")
print(f"  Test  images: 67 (unchanged)")
```

```python
import os, json, numpy as np
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV3Small
from tensorflow.keras.applications.mobilenet_v3 import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# ── PATHS ─────────────────────────────────────────────────────────────────────
TRAIN_DIR = "dataset/train_augmented"
VALID_DIR = "dataset/valid"
TEST_DIR  = "dataset/test"
CLASSES   = ["Black Soil", "Laterite Soil", "Yellow Soil"]
IMG_SIZE  = (224, 224)
BATCH     = 32

# ── DATA LOADERS ──────────────────────────────────────────────────────────────
# No augmentation here — already done on disk
# Just preprocess_input (scales to [-1, 1] as MobileNetV3 expects)

train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input
).flow_from_directory(
    TRAIN_DIR, target_size=IMG_SIZE,
    batch_size=BATCH, class_mode="categorical", shuffle=True
)

val_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input
).flow_from_directory(
    VALID_DIR, target_size=IMG_SIZE,
    batch_size=BATCH, class_mode="categorical", shuffle=False
)

test_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input
).flow_from_directory(
    TEST_DIR, target_size=IMG_SIZE,
    batch_size=BATCH, class_mode="categorical", shuffle=False
)

# Save class index for inference
with open("class_indices.json", "w") as f:
    json.dump(train_gen.class_indices, f)
print("Class map:", train_gen.class_indices)

# ── BUILD MODEL ───────────────────────────────────────────────────────────────
base = MobileNetV3Small(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3),
    include_preprocessing=False   # we handle preprocessing ourselves
)
base.trainable = False            # freeze for phase 1

x = base.output
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dense(256, activation="relu")(x)
x = Dropout(0.4)(x)
x = Dense(128, activation="relu")(x)
x = Dropout(0.3)(x)
out = Dense(len(CLASSES), activation="softmax")(x)

model = Model(inputs=base.input, outputs=out)
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
model.summary()

# ── PHASE 1: Train head only (base frozen) ────────────────────────────────────
print("\n🔵 Phase 1: Training head only...")

callbacks_p1 = [
    EarlyStopping(monitor="val_loss", patience=5,
                  restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                      patience=3, verbose=1),
    ModelCheckpoint("best_soil_model.h5", monitor="val_accuracy",
                    save_best_only=True, verbose=1),
]

history1 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=15,
    callbacks=callbacks_p1
)

# ── PHASE 2: Fine-tune last 40 layers ────────────────────────────────────────
print("\n🟠 Phase 2: Fine-tuning last 40 layers...")

for layer in base.layers[-40:]:
    layer.trainable = True

# Much lower LR to avoid destroying pretrained weights
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

callbacks_p2 = [
    EarlyStopping(monitor="val_loss", patience=5,
                  restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                      patience=3, verbose=1),
    ModelCheckpoint("best_soil_model.h5", monitor="val_accuracy",
                    save_best_only=True, verbose=1),
]

history2 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,
    callbacks=callbacks_p2
)

# ── EVALUATE ON TEST SET ──────────────────────────────────────────────────────
print("\n🧪 Evaluating on held-out test set...")
model.load_weights("best_soil_model.h5")   # load best checkpoint

loss, acc = model.evaluate(test_gen, verbose=0)
print(f"\n✅ Test accuracy: {acc*100:.2f}%")

preds      = np.argmax(model.predict(test_gen), axis=1)
idx_to_cls = {v: k for k, v in train_gen.class_indices.items()}
pred_lbls  = [idx_to_cls[i] for i in preds]
true_lbls  = [idx_to_cls[i] for i in test_gen.classes]
print("\nPer-class report:")
print(classification_report(true_lbls, pred_lbls))

# ── PLOT TRAINING CURVES ──────────────────────────────────────────────────────
def plot_history(h1, h2):
    acc  = h1.history["accuracy"]      + h2.history["accuracy"]
    vacc = h1.history["val_accuracy"]  + h2.history["val_accuracy"]
    loss = h1.history["loss"]          + h2.history["loss"]
    vloss= h1.history["val_loss"]      + h2.history["val_loss"]
    ep   = range(1, len(acc) + 1)
    p1_end = len(h1.history["accuracy"])  # mark phase boundary

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    ax1.plot(ep, acc,  label="Train acc")
    ax1.plot(ep, vacc, label="Val acc")
    ax1.axvline(p1_end, color="gray", linestyle="--", label="Phase 2 start")
    ax1.set_title("Accuracy"); ax1.legend()

    ax2.plot(ep, loss,  label="Train loss")
    ax2.plot(ep, vloss, label="Val loss")
    ax2.axvline(p1_end, color="gray", linestyle="--", label="Phase 2 start")
    ax2.set_title("Loss"); ax2.legend()

    plt.tight_layout()
    plt.savefig("training_curves.png", dpi=120)
    plt.show()
    print("✅ Training curves saved!")

plot_history(history1, history2)
model.save("soil_classifier_final.h5")
print("✅ Final model saved!")


# ── INFERENCE WITH EDGE CASE HANDLING ────────────────────────────────────────
CONFIDENCE_THRESHOLD = 0.65   # below this → reject

with open("class_indices.json") as f:
    class_indices = json.load(f)
idx_to_cls = {v: k for k, v in class_indices.items()}

def predict_soil(image_path):
    """
    Returns:
        soil_type (str or None)
        confidence (float)
        error_message (str or None)
    """
    # Load & preprocess
    img = tf.keras.utils.load_img(image_path, target_size=IMG_SIZE)
    arr = tf.keras.utils.img_to_array(img)
    arr = preprocess_input(arr)
    arr = np.expand_dims(arr, axis=0)

    preds = model.predict(arr, verbose=0)[0]
    idx   = np.argmax(preds)
    conf  = float(preds[idx])
    label = idx_to_cls[idx]

    # Edge case 1: wrong image / not soil at all
    # Edge case 2: unknown 4th soil type
    # Both handled by the same threshold check
    if conf < CONFIDENCE_THRESHOLD:
        return None, conf, (
            f"⚠️ Could not confidently identify the soil type "
            f"(confidence: {conf:.0%}). "
            f"This may be an unrecognized soil type or an invalid image. "
            f"Please upload a clear close-up photo of Black, Laterite, "
            f"or Yellow soil, or enter soil parameters manually."
        )

    return label, conf, None


# ── TEST THE INFERENCE FUNCTION ───────────────────────────────────────────────
test_image = os.path.join(TEST_DIR,
             CLASSES[0],
             os.listdir(os.path.join(TEST_DIR, CLASSES[0]))[0])

soil, conf, err = predict_soil(test_image)

if err:
    print(err)
else:
    print(f"🪨  Predicted: {soil} ({conf:.0%} confidence)")
```
